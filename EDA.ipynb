```python
import json


with open("data/europe_data.json") as f:
    data = json.load(f)
```

```python
exists_countries = []

for country in data['countries']:
    exists_countries.append(country['name'].lower())
print(len(exists_countries))
# print(exists_countries)

new_countries = []
for country in countries:
    if country not in exists_countries:
        new_countries.append(country)
```

```python
import google.generativeai as genai

GOOGLE_API_KEY = ""
genai.configure(api_key=GOOGLE_API_KEY)
llm = genai.GenerativeModel('models/gemini-2.5-flash')

# response = llm.generate_content(prompt)
# return response.text

def get_information_from_llm(country):
    prompt = f"""
    You are an expert about geography and you know about the information of every countries around the world.
    So, you help me fill the information for the country: {country} based on the json structre like below. And, you have to answer with a json-style response.
    For example: 
    Country: Poland
    The json structure:
    {{"name": "Poland", "eu_member": null, "capital": "Warsaw", "borders_with": ["Belarus", "Czech Republic", "Germany", "Lithuania", "Russia", "Slovakia"]}}
"""
    
    response = llm.generate_content(prompt)
    return response.text


results = []
for country in new_countries:
    result = get_information_from_llm(country)
    json_result = json.loads("\n".join(result.split("\n")[1:-1]))
    data['countries'].append(json_result)

print("The number of countries in Europe: ", len(data['countries']))
```

```python
with open("data/updated_europe_data.json") as f:
    data = json.load(f)

for country in data['countries']:
    if country['name'].lower() not in countries:
        print(country)
```

```python
def get_information_from_llm(prompt):
    
    response = llm.generate_content(prompt)
    return response.text
```

```python
DATA = data
for i, country in enumerate(DATA['countries']):
    prompt = f"""You are an expert about geography. Now, you help me answer the question: 
    Is {country['name']} in EU organization or not?
    You have to answer Yes or No, do not add any information.
    For example:
    Question: Is Germany in EU organization or not?
    Answer: Yes
    Question: Is {country} in EU organization or not?
    Answer: """
    answer = get_information_from_llm(prompt)
    DATA['countries'][i]['eu_member'] = answer
```

```python
with open("data/updated_europe_data.json", "w") as f:
    json.dump(DATA, f)
```
